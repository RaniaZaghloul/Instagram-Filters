<h2> 2. Script

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
import numpy as np
%matplotlib inline
def insta_like(image,instafilter):
    if (instafilter == 'Moon'):
        output = Moon(image)
    elif (instafilter == 'Lark'):
        output = Lark(image)
    elif (instafilter == 'Aden'):
        output = Aden(image)
    elif (instafilter == '1997'):
        output = f1997(image)
    elif (instafilter == 'EarlyBird'):
        output = EarlyBird(image)
    return output
```

<h2> 2.0 Moon Filter

```python
def Moon(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    cv2.imwrite('gray_image.png',gray_image)
    gray = cv2.imread('gray_image.png')
    HSV=cv2.cvtColor(gray,cv2.COLOR_BGR2HSV)
    HSV[:,:,0] += 300 #color
    HSV[:,:,2] -= 3 #BRIGHTNESS
    img =cv2.cvtColor(HSV,cv2.COLOR_HSV2RGB)
    return img
```

<h2> 2.1 Lark Filter

```python
def Lark(image):
    clahe = cv2.createCLAHE(clipLimit=3., tileGridSize=(8,8))
    lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)  
    ch1, ch2, ch3 = cv2.split(lab)  
    ch1_edited = clahe.apply(ch1)
    lab = cv2.merge((ch1_edited,ch2,ch3))  
    img = cv2.cvtColor(lab, cv2.COLOR_LAB2BGR) 
    return img
```

<h22.2 Aden Filter

```python
def Aden(image):
    HSV=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
    HSV[:,:,0] += 75 #color                    
    HSV[:,:,1] += 5 #SATURATION
    HSV[:,:,2] += 2 #BRIGHTNESS
    img =cv2.cvtColor(HSV,cv2.COLOR_HSV2RGB)
    return img
```

2.3 1997 Filter

```python
def f1997(image):
    HSV=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
    HSV[:,:,0] = 0 #color
    HSV[:,:,1] +=2 #SATURATION
    img =cv2.cvtColor(HSV,cv2.COLOR_HSV2RGB)
    return img
```

2.4 EarlyBird

```python
def EarlyBird(image):
    HSV=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
    HSV[:,:,0] = 30 #color
    y=cv2.cvtColor(HSV,cv2.COLOR_HSV2RGB)
    HSV=cv2.cvtColor(y,cv2.COLOR_BGR2HSV)
    HSV[:,:,0] += 100 #color
    img =cv2.cvtColor(HSV,cv2.COLOR_HSV2RGB)
    return img
```
